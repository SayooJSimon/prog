```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data =  pd.read_csv("C:\\Users\\shadr\\OneDrive\\Desktop\\Iris DataSet\\Iris.csv")
x = data['petal.length']
y = data['petal.width']
plt.title('Iris Dataset')
plt.ylabel('PetalLenghtCm')
plt.xlabel('PetalWidthCm')
np.where(['PetalLenghtCm']>['PetalWidthCm'])
plt.scatter(x,y,c = data['petal.length'],s = data['petal.width'],marker ='D', alpha=0.5)
plt.colorbar(label = 'Petal Lenght(cm)')
plt.show()
```
